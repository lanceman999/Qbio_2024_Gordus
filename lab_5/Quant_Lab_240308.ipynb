#  Specificity and Occupancy

For this lab, you will calculate specificity and occupancy based on your calculations from the worksheet.



```python
# These are the libraries you will need for this lab

import matplotlib.pyplot as plt
import numpy as np
```


# 1: Variable assignments

Create and assign values for the following variables:

**Scalars:**

$K_{D,Op}$

$K_{D,nonOp}$

$N_{A}$

*Cell Volume*

$[Op]_{Total}$

$[nonOp]_{Total}$

**Vector:**

*LacR*: This variable should span integers 1 through 10

# <font color='red'>Student Input Cell: 4 points</font>   

```python
KdOP = 10**(-12)
KdnonOP = 10**(-4)
Na = 6.023 * (10**23)
CV = 1.6*(10**(-15))
OP_total = (1/Na)*(1/CV)
nonOP_total = ((4.5*(10**6))/Na)*(1/CV)
LacR = np.arange(1,11)
```

# 2: Functions

Use the variables above to define the following vectors as functions of LacR:

$[LacR]_{Total}$

$[LacR\cdot nonOp]$

$[LacR]_{Free}$

$[LacR\cdot Op]$

*Op occupancy*

*Op alpha*


# <font color='red'>Student Input Cell: 12 points</font>   

```python
LacR_total = (LacR/Na)*(1/CV)
LacR_nonop = (LacR_total*nonOP_total)/(KdnonOP+nonOP_total)
LacR_free = LacR_total - LacR_nonop
LacR_OP = (LacR_free*OP_total)/(KdOP+LacR_free)
OP_occ = LacR_OP/OP_total
OP_alpha = LacR_OP/LacR_nonop
```

# 3: Graph Results

Plot 2 Graphs:

*occupancy* vs *LacR*

*alpha* vs *LacR*

**NOTE:** The plots should be DISCRETE, i.e. *LacR* is a vector of integers, so you should plot discrete dots, not continuous lines.

# <font color='red'>Student Input Cell: 4 points</font>   

```python
plt.scatter(LacR, OP_occ)

plt.title("Occupancy vs LacR")

plt.xlabel("LacR")
plt.ylabel("Occupancy")

plt.show()



plt.scatter(LacR, OP_alpha)

plt.xlabel("LacR")
plt.ylabel("Occupancy")
plt.title("Alpha vs LacR")

plt.tight_layout()
plt.show()
```
