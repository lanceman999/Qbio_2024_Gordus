**Do the following:**

A protein has two binding sites. The Kd for binding the first ligand is 10<sup>-6</sup>. Plot the Fraction bound vs. concentration of ligand (log scale) if the cooperativity ($\alpha$) between the first and second binding sites is a) 100, b) 1, or c) 0.01.

## Libraries you will need

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Student Input Cell

```python
### Defining variables & equations ###
kd1 = 10**-6

alpha100 = 100
alpha1 = 1
alpha01 = 0.01

Ligand_conc = np.linspace(10**-3, 10**-10, 1000000)

def cooperative_binding(alpha):
    Q = 1 + 2*Ligand_conc/(kd1) + Ligand_conc**2/(kd1**2*alpha)
    T = 1/Q
    TL = (2*Ligand_conc / kd1) / Q
    TL2 = (Ligand_conc**2 / (kd1**2 * alpha)) / Q
    Y = (TL/2 + TL2) / (T + TL + TL2)
    return Y

results_100 = cooperative_binding(alpha100)
results_1 = cooperative_binding(alpha1)
results_01 = cooperative_binding(alpha01)
```

```python
### Plotting ###
plt.plot(np.log10(Ligand_conc),results_100, color = 'purple', label = "alpha = 100")
plt.plot(np.log10(Ligand_conc),results_1, color = 'red', label = "alpha = 1")
plt.plot(np.log10(Ligand_conc),results_01, color = 'cyan', label = "alpha = 0.01")

plt.title("Concentration of Ligand vs Fraction Bound")
plt.ylabel("Fraction Bound") #of ligand bound to protein
plt.xlabel("log [ligand]")
```
